```python
#Primero cargamos los paquetes que necesitamos para trabajar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
#Leemos el archivos
df = pd.read_csv("sargento_201117.csv", header=None, skiprows=1, names=['#','DateTime', 'Pres', 'Temp','WaterLvl'])
df.head()

```

```python
# Convertir la cadena de caracteres 'DateTime' en variable temporal 'NDate'
df['Ndate'] = pd.to_datetime(df['DateTime'], format='%m %d %Y %H:%M:%S')
df['month'] = df['Ndate'].dt.month
df.head()
```

```python
#Hacemos la grafica de caja con Seaborn
ax = sns.boxplot(x="month", y="WaterLvl", data=df)
plt.show()
```

```python
#Tambien ocupamos leer el archivo de salinidad, por lo que tenemos que cargarlo.
dg = pd.read_csv("sargento-salinidad-201117.csv", header=None, skiprows=2, names=['#','DateTime','Cond','Temp','SpecCond','Sal'])
dg.head()
```

```python
# Convertir la cadena de caracteres 'DateTime' en variable temporal 'NDate'
dg['Ndate'] = pd.to_datetime(dg['DateTime'], format='%m %d %Y %H:%M:%S')
dg['month'] = dg['Ndate'].dt.month
dg.head()
```

```python
#Hacemos la grafica de caja con Seaborn
ax = sns.boxplot(x="month", y="Sal", data=dg)
plt.show()
```

```python
#Hacemos la grafica de caja con Seaborn
ax = sns.boxplot(x="month", y="Temp", data=dg)
plt.show()
```

```python
# Realiza un análisis exploratorio de datos de df
df.describe()
```

```python
# Realiza un análisis exploratorio de datos de dg
dg.describe()
```

```python
#Ahora ocupamos concatenar datos para tener un solo dataframe 
#Este dataframe solamente tiene lo que necesitamos para la siguiente parte de la actividad.
dh=pd.concat([df['#'],df['DateTime'],df['WaterLvl'],dg['Sal'],dg['Temp']], axis=1)
dh.head()
```

```python
#Exploraremos si existe correlacion entre parejas de variables de este nuevo dataframe
#Empezamos con Nivel del Mar y Salinidad
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("WaterLvl", "Sal", data=dh, kind="reg",
                   color="g", size=7)
plt.show(g)
```

```python
#Luego Nivel del Mar y Temperatura
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("WaterLvl", "Temp", data=dh, kind="reg",
                   color="b", size=7)
plt.show(g)
```

```python
#Por ultimo, Salinidad con Temperatura
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("Sal", "Temp", data=dh, kind="reg",
                   color="y", size=7)
plt.show(g)
```

```python
#El siguiente paso nos dice que con ayuda de Matplotlib hagamos graficas individuales en función del tiempo.
#primero Nivel del Mar
plt.plot_date(x=dh.DateTime, y=dh.WaterLvl, fmt="g")
plt.title("Variación del nivel del mar contra fecha")
plt.ylabel("Metros (mts)")
plt.grid(True)
plt.show()
```

```python
#Ahora salinidad.
plt.plot_date(x=dh.DateTime, y=dh.WaterLvl, fmt="b")
plt.title("Variación de la Salinidad contra fecha")
plt.ylabel("Partes por mil (ppt)")
plt.grid(True)
plt.show()
```

```python
#Por ultimo, una gráfica de Temperatura contra Fecha
plt.plot_date(x=dh.DateTime, y=dh.Temp, fmt="y")
plt.title("Variación de la Temperatura contra fecha")
plt.ylabel("Temperatura (C°)")
plt.grid(True)
plt.show()
```

```python
#Perfecto, hasta ahora todo fine, seguimos.
#Esta parte nos pide hacer graficas superpuestas con doble eje vertical. 
#Primero es de Nivel del Mar con Salinidad
dh1 = dh[['WaterLvl','Sal']]
plt.figure(); dh1.plot(); plt.legend(loc='best')
plt.title("Variación del nivel del mar y de la Salinidad")
plt.ylabel("Metros (mts) / Partes por mil (ppt)")
plt.grid(True)
plt.show()
```

```python
#Nivel del mar y Temperatura
plt.plot_date(x=dh.DateTime, y=dh.WaterLvl, fmt="b")
plt.plot_date(x=dh.DateTime, y=dh.Temp, fmt="g")
plt.title("Variación del Nivel del Mar y Temperatura")
plt.ylabel("Metros (mts) / Temp (C°)")
plt.grid(True)
plt.show()
```

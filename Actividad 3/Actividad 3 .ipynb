```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv('Albu_Jun.txt', skiprows=2, sep='\s+')
```

```python
# Dar estructura de datos (DataFrame)
df = pd.DataFrame(df0)
```

```python
#Las gráficas de dos variables en función de otra se declarabn de manera idependiente cada uno.
#Donde "fmt" es el color de la función y "b-" es blue, "r-" es red, y así.
plt.plot_date(x=df.HGHT, y=df.TEMP, fmt="b-")
plt.plot_date(x=df.HGHT, y=df.DWPT, fmt="y-")
plt.title("Temperatura y Temperatura de Rocio como función de Altura.")
plt.ylabel("Temperatura en C°")
plt.grid(True)
plt.show()
```

```python
plt.plot_date(x=df.HGHT, y=df.SKNT, fmt="b-")
plt.title("Variación de rapidez en nudos como función de altura")
plt.ylabel("Rapidez en nudos")
plt.grid(True)
plt.show()
```

```python
plt.plot_date(x=df.HGHT, y=df.RELH, fmt="g-")
plt.title("Variación de humedad relativa como función de altura.")
plt.ylabel("Porcentaje")
plt.grid(True)
plt.show()
```
